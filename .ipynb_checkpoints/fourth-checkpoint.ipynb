```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
import numpy as number
import squarify
```

```python
# import os
funding = pd.read_csv("D:\MCA\sem4\DS\india_funding1.csv")
# before print all coumns and show bollean
print(funding.isnull().sum().sort_values(ascending =False))
funding.Industry.fillna("no data", inplace=True)
funding.SubVertical.fillna("no record", inplace=True)
funding.Location.fillna("location not found", inplace=True)
funding.Investors.fillna("investor is no", inplace=True)
funding.InvestmentType.fillna("type isn't display",inplace=True)
funding.Amount.fillna("Amount blank",inplace=True)

fund_count  = funding['StartupName'].value_counts()
fund_count = fund_count.head(40)
# print(fund_count)
plt.figure(figsize=(15,8))
sn.barplot(fund_count.index, fund_count.values, alpha=0.9)
plt.xticks(rotation='vertical')
plt.xlabel('Startups', fontsize=12)
plt.ylabel('Number of fundings made', fontsize=12)
plt.title("Startups-Number of fundings distribution", fontsize=16)
plt.show()
```
