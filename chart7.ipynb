```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sn
import squarify
# What are different types of funding for startups ?
funding = pd.read_csv("D:\MCA\sem4\DS\india_funding1.csv")
investment = funding['InvestmentType'].value_counts()
# print(investment)
plt.figure(figsize=(19,8))
sn.barplot(investment.index, investment.values, alpha=0.6)
plt.xticks(rotation='vertical')
plt.xlabel('Investment Type', fontsize=12)
plt.ylabel('Number of fundings made', fontsize=12)
plt.title("Investment Type with number of funding", fontsize=16)
plt.show()
```
